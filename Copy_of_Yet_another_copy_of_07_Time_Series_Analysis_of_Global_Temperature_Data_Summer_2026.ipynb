### Question 1: Import Libraries and Load Data (2 Marks)

Import `pandas` as `pd` and `numpy` as `np`. Download the global temperature dataset `monthly_csv.csv` from https://drive.google.com/drive/folders/1gieHICVDBbUKMZiSF4YRQMAJic-w50JM?usp=sharing and load into a pandas DataFrame called `df_temp`. Display the first 5 rows.


```python
import pandas as pd
import numpy as np
df_temp = pd.read_csv('monthly_csv.csv')
df_temp.head()
```

### Question 2: Convert to Datetime and Sort (1 Mark)

The 'Date' column is currently a string. Convert it to a proper datetime object and then sort the entire DataFrame `df_temp` by this 'Date' column in ascending order. Display the head of the sorted DataFrame.

**Expected Output:** A table showing the first 5 rows of the dataset, now sorted by date starting from the earliest records.

```python
import pandas as pd

df_temp['Date'] = pd.to_datetime(df_temp['Date'])

df_temp = df_temp.sort_values(by='Date', ascending=True)

print(df_temp.head())
```

### Question 3: Data Inspection (2 Marks)

Get a quick overview of your sorted DataFrame. Print the shape and then check for the total number of missing (null) values across the entire DataFrame.

**Expected Output:** The shape of the DataFrame and the total count of null values.

```python

print("Shape of DataFrame:", df_temp.shape)

total_nulls = df_temp.isnull().sum().sum()
print("Total Missing Values:", total_nulls)
```

### Question 4: Extract Year and Month (1 Marks)

Create two new columns in `df_temp`: 'Year' and 'Month'. These should contain the year and month numbers extracted from the 'Date' column. Display the head of the DataFrame to show these new columns.

**Expected Output:** The first 5 rows of the DataFrame, now including 'Year' and 'Month' columns.

```python

df_temp['Year'] = df_temp['Date'].dt.year
df_temp['Month'] = df_temp['Date'].dt.month

print(df_temp.head())
```

### Question 5: Calculate 12-Month Moving Average (2 Marks)

Calculate the 12-month moving average of the 'Mean' temperature for each 'Source'. Store this in a new column named `Moving_Avg`. Display the last 5 rows of the DataFrame.

**Expected Output:** A table showing the last 5 rows of the DataFrame with the new 'Moving_Avg' column.

```python

df_temp['Moving_Avg'] = (
    df_temp.groupby('Source')['Mean']
           .transform(lambda x: x.rolling(window=12).mean())
)

print(df_temp.tail())
```

### Question 6: Filter for the Last 20 Years (2 Marks)

Create a new DataFrame called `df_last20` by filtering `df_temp` to include only data from the last 20 years (inclusive of the latest year in the data). Print the shape of this new DataFrame.

**Expected Output:** A tuple representing the shape of the filtered `df_last20` DataFrame.

```python

latest_year = df_temp['Year'].max()

df_last20 = df_temp[df_temp['Year'] >= latest_year - 19]

print(df_last20.shape)
```

### Question 7: Group by Source (2 Marks)

Group the original `df_temp` DataFrame by the 'Source' column and calculate the mean of the 'Mean' temperature for each source. Display the resulting Series.

**Expected Output:** A Pandas Series showing the two sources (GCAG, GISTEMP) as the index and their overall average mean temperature as the values.

```python

source_avg = df_temp.groupby('Source')['Mean'].mean()

print(source_avg)
```

### Question 8: Aggregate by Year (1 Marks)

Group the `df_temp` DataFrame by 'Year' and calculate the average 'Mean' temperature for each year. Display the last 5 entries of this yearly aggregation.

**Expected Output:** A Series showing the last 5 years in the dataset and their corresponding average mean temperatures.

```python

yearly_avg = df_temp.groupby('Year')['Mean'].mean()

print(yearly_avg.tail())
```

### Question 9: Find the Hottest Month on Average (2 Marks)

Across all years in the `df_temp` dataset, find which month has the highest average 'Mean' temperature. Print the resulting Series which shows the average temperature for each month.

**Expected Output:** A Series with month numbers (1-12) as the index and their corresponding average mean temperatures.

```python

monthly_avg = df_temp.groupby('Month')['Mean'].mean()

print(monthly_avg)
```

### Question 10: Create a Pivot Table (2 Marks)

Using the `df_last20` DataFrame, create a pivot table to show the average 'Mean' temperature for each 'Month' across each 'Year'. Fill any missing values with 0. Display the resulting pivot table.

**Expected Output:** A table with years as rows, months (1-12) as columns, and the average mean temperature as the values.

```python

pivot_temp = pd.pivot_table(
    df_last20,
    values='Mean',
    index='Year',
    columns='Month',
    aggfunc='mean',
    fill_value=0
)

print(pivot_temp)
```

### Question 11: Load MNIST Data (3 Marks)

Load the MNIST dataset using `sklearn.datasets.load_digits`. Separate the dataset into features (`X`) and target labels (`y`).
Print the shape of the features and the target arrays.

**Expected Output:** The shape of `X` and `y`.

```python
from sklearn.datasets import load_digits

digits = load_digits()

X = digits.data
y = digits.target

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

### Question 12: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10, random_state=42)

kmeans_labels = kmeans.fit_predict(X)

print(kmeans_labels)

print("Shape of cluster labels:", kmeans_labels.shape)
```

### Question 13: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
from scipy.stats import mode
from sklearn.metrics import f1_score
import numpy as np

mapped_labels = np.zeros_like(kmeans_labels)

for i in range(10):
    mask = (kmeans_labels == i)
    if np.sum(mask) > 0:
        mapped_labels[mask] = mode(y[mask], keepdims=False).mode

f1 = f1_score(y, mapped_labels, average='macro')

print("Macro F1 Score:", f1)
```
